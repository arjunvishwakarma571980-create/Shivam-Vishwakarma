#### 1) Instruction (For Data Cleaning)
 Find all Null Value in the Dataset(if there is null value in any column, then fill it with the mean of that column

```python
df.isnull().sum()
```

```python
df.fillna({"Make": df["Make"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Model": df["Model"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Type": df["Type"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Origin": df["Origin"].mode()[0]}, inplace=True)
```

```python
df.fillna({"DriveTrain": df["DriveTrain"].mode()[0]}, inplace=True)
```

```python
df.fillna({"MSRP": df["MSRP"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Invoice": df["Invoice"].mode()[0]}, inplace=True)
```

```python
df.fillna({"EngineSize": df["EngineSize"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Cylinders": df["Cylinders"].mean()}, inplace=True)
```

```python
df.fillna({"Horsepower": df["Horsepower"].mode()[0]}, inplace=True)
```

```python
df.fillna({"MPG_City": df["MPG_City"].mode()[0]}, inplace=True)
```

```python
df.fillna({"MPG_Highway": df["MPG_Highway"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Weight": df["Weight"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Wheelbase": df["Wheelbase"].mode()[0]}, inplace=True)
```

```python
df.fillna({"Length": df["Length"].mode()[0]}, inplace=True)
```
