### Q15) Identify departments with the highest attrition rate (Resigned %)

```python
df.head(2)
```

```python
dept_counts = df.groupby("Department")["Status"].agg(total_emp = "count", resigned= lambda x: (x == "Resigned").sum())

dept_counts
```

```python
type(dept_counts)
```

```python
#Calculate the attribution rate

dept_counts["attribution_rate_%"] = (dept_counts["resigned"] / dept_counts["total_emp"]) * 100
```

```python
dept_counts
```

```python
#sort by attrition rate (highest first)

dept_counts.sort_values("attribution_rate_%", ascending = False)
```
